# Задание 1

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
bost = load_boston()
```

```python
bost.keys()
```

```python
data = bost.data
```

```python
targ = bost.target
```

```python
X = pd.DataFrame(data, columns=bost.feature_names)
```

```python
X.head()
```

```python
y = pd.DataFrame(targ, columns=['price'])
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
score1 = r2_score(y_test, y_pred)
```

```python
score1
```

# Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(random_state=42, max_depth=12, n_estimators=1000)
```

```python
model.fit(X_train, y_train.price)
```

```python
y_pred = model.predict(X_test)
```

```python
score2 = r2_score(y_test, y_pred)
```

```python
score2
```

```python
print(f'Model 1 = {score1 * 100:0.2f}%\nModel 2 = {score2 * 100:0.2f}%')
```

#### Вывод
Видно, что 2 модель построенная на RandomForestRegressor превосходит LinearRegression на 16%, что значительно

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split

```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.V1.value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
y = df.Class
```

```python
df = df.drop('Class', axis=1)
```

```python
df.head()
```

```python
X = df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=100, stratify=y)
```

```python
X_train.describe()
```

```python
y_train.describe()
```

```python
parameters = [{'n_estimators': [10, 15], 
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
mod = GridSearchCV(estimator=RandomForestClassifier(random_state=100), 
param_grid=parameters,
scoring='roc_auc',
cv=3)
```

```python
mod.fit(X_test, y_test)
```

```python
mod.best_params_
```

```python
mod = RandomForestClassifier(max_depth=4, max_features=4, n_estimators=15, random_state=100)
mod.fit(X_test, y_test)
y_pred = mod.predict(X_test)
```

```python
y_pred_proba = mod.predict_proba(X_test)
```

```python
y_pred_proba = y_pred_proba[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
